## 1.Is strike rate significantly higher for high-run scorers?
### Ho : Mean strike rate is equal
### H1 : Mean strike rate is differs

```python
from scipy import stats

m = df['Runs'].mean()
group_high = df[df['Runs']>=m]['Strike_rate']
group_low = df[df['Runs']<m]['Strike_rate']
t_stat, p_value = stats.ttest_ind(group_high, group_low, equal_var=True)
print(f'T-stat:{t_stat:.4f} and P-value {p_value:.4f}')
if p_value< 0.05:
    print("Conclusion : There is a statistically significant difference in Strike Rate")
else:
    print("Conclusion : There is no Statistically significant difference in Strike Rate")
print('-' * 140)
```

## 2.Is strike rate correlated with total runs?
### Ho : No correlation
### H1 : Significant correlation

```python
corr_coef, p_value = stats.pearsonr(df['Runs'],df['Strike_rate'])
print(f'Pearson Correlation Coefficient:{t_stat:.4f} and P-value {p_value:.4f}')
if p_value< 0.05:
    print("Conclusion: Reject H. There is a statistically significant correlation "
          "between Strike Rate and Total Runs.")
else:
    print("Conclusion: Fail to reject H. There is no statistically significant correlation "
          "between Strike Rate and Total Runs.")
print('-' * 140)
```

## 3.Is ball faced correlated with runs scored?
### Ho : No correlation
### H1 : Significant correlation

```python
scc, p_value = stats.spearmanr(df['Balls_faced'], df['Runs'])
print(f"Spearman Correlation Coefficient (): {scc:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusion: Reject H. There is a statistically significant relationship between Balls Faced and Runs")
else:
    print("Conclusion: Fail to reject H. There is no statistically significant relationship between Balls Faced and Runs")
print('-' * 140)
```

## 4.Do players with higher strike rates face fewer balls?
### Ho : Mean ball faced is equal
### H1 : Mean ball faced is different

```python
m = df['Strike_rate'].median()
group_high = df[df['Strike_rate'] >= m]['Balls_faced']
group_low = df[df['Strike_rate'] < m]['Balls_faced']

t_stat, p_value = stats.ttest_ind(group_high, group_low, equal_var=False)

print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject H. There is a statistically significant difference "
          "in Balls Faced between players with Strike Rate  median and < median.")
else:
    print("Conclusion: Fail to reject H. There is no statistically significant difference "
          "in Balls Faced between players with Strike Rate  median and < median.")
print('-' * 140)
```

## 5.Is highest score significantly higher for high-average batters?
### H: Mean highest score is equal
### H: Mean highest score differs

```python
m = df['Average'].median()
group_high = df[df['Average']>= m]['Highest_score']
group_low = df[df['Average'] <m]['Highest_score']

u_stat, p_value = stats.mannwhitneyu(group_high, group_low, alternative='two-sided')
print(f"MannWhitney U statistic: {u_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusion: Reject H. The highest score of high-average batters "
          "is significantly different from that of low-average batters.")
else:
    print("Conclusion: Fail to reject H. There is no significant difference "
          "in highest scores between high- and low-average batters.")
print('-' * 140)
```

## 7. Is run consistency different for high vs low average batters?
### Ho : Equal variance
### H1 : Variance differs

```python
m = df['Average'].median()

group_high = df[df['Average'] >= m]['Runs']
group_low = df[df['Average'] < m]['Runs']
levene_stat, p_value = stats.levene(group_high, group_low)

print(f"Levene Statistic: {levene_stat:.4f}")
print(f"P-value: {p_value:.4f}")
if p_value < 0.05:
    print("Conclusion: Reject H. Run consistency is significantly different between high- and low-average batters.")
else:
    print("Conclusion: Fail to reject H. There is no significant difference in run consistency between high- and low-average batters.")
print('-' * 140)
```

```python
six_ratio = df['Number_of_6s'] / (df['Number_of_4s'] + 1)
m = six_ratio.median()
group_high = df[six_ratio >= m]['Strike_rate']
group_low = df[six_ratio < m]['Strike_rate']

t_stat, p_value = stats.ttest_ind(group_high, group_low, equal_var=False)
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Conclusion: Reject H. Strike rates differ significantly between batters with high and low six-to-four ratios.")
else:
    print("Conclusion: Fail to reject H. There is no statistically significant difference in strike rates between batters with high and low six-to-four ratios.")
print('-'* 140)
```

##  17.Do individual players have significantly different boundary rates?

### H: Boundary rate is equal across players

### H: Boundary rate differs across players

```python
df['boundary_rate'] = (df['Number_of_4s'] + df['Number_of_6s']) / df['Balls_faced']
filtered = df[df['Balls_faced'] >= 100]

groups = [g['boundary_rate'] for _, g in filtered.groupby('Player')]
stat, p_value = stats.kruskal(*groups)

print(f'Kruskal-Wallis H-Statistics : {stat}')
print(f'P-value : {p_value}')

if p_value < 0.05:
    print("The boundary rates differ significantly across players.")
else:
    print("The boundary rates do not differ significantly across players.")
print('-'*140)
```

## 19 .Are the batting averages of players with low, medium, and high boundary counts equal?

### Ho: The batting averages of players in the low, medium, and high boundary groups are equal.

### H1: The batting averages of players in the low, medium, and high boundary groups are not equal

```python
boundary_quantiles = df['boundaries'].quantile([0.33, 0.66])
low_boundary = df['boundaries'] < boundary_quantiles.iloc[0]
high_boundary = df['boundaries'] >= boundary_quantiles.iloc[1]
medium_boundary = ~(low_boundary | high_boundary)

df['boundary_group'] = 'Medium'
df.loc[low_boundary, 'boundary_group'] = 'Low'
df.loc[high_boundary, 'boundary_group'] = 'High'

print(df[['Player','boundaries','boundary_group']].head())

low_boundary_avg = df[df['boundary_group'] =='Low']['Average']
medium_boundary_avg = df[df['boundary_group'] == 'Medium']['Average']
high_boundary_avg = df[df['boundary_group'] == 'High']['Average']
```

```python
h_stat, p_value = stats.kruskal(low_boundary_avg, medium_boundary_avg, high_boundary_avg)
print(f"Kruskal-Wallis H-statistic: {h_stat}")
print(f"P-value: {p_value}")
if p_value < 0.05:
    print("The batting averages differ significantly between boundary groups.")
else:
    print("The batting averages do not differ significantly between boundary groups.")
print("-"*140)
```

## 20.Do strike rates differ significantly across players with different numbers of hundreds scored?

### Ho: The strike rate is equal across the three groups based on number of 100s scored
### H1 : The strike rate differs significantly across the three groups based on the number of 100s scored.

```python
group_0 = df['Number_of_100'] == 0
group_1 = df['Number_of_100'] == 1
group_more_than_1 = df['Number_of_100'] > 1

df['hundred_group'] = '0 hundreds'  # Default group
df.loc[group_1, 'hundred_group'] = '1 hundred'
df.loc[group_more_than_1, 'hundred_group'] = 'More than 1 hundred'

print(df[['Player', 'Number_of_100', 'hundred_group']].head())

group_0_strike_rate = df[df['hundred_group'] == '0 hundreds']['Strike_rate']
group_1_strike_rate = df[df['hundred_group'] == '1 hundred']['Strike_rate']
group_more_than_1_strike_rate = df[df['hundred_group'] == 'More than 1 hundred']['Strike_rate']

h_stat, p_value = stats.kruskal(group_0_strike_rate, group_1_strike_rate, group_more_than_1_strike_rate)
print(f"Kruskal-Wallis H-statistic: {h_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("The strike rates differ significantly across the number of hundreds groups.")
else:
    print("The strike rates do not differ significantly across the number of hundreds groups.")
print('-'*140)
```
